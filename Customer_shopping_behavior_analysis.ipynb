```python

# Carregando o dataset usando pandas
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
# Estatísticas resumidas usando .describe()
df.describe(include='all')
```

```python
# Verificando se há dados ausentes ou valores nulos no conjunto de dados
df.isnull().sum()
```

```python
# Imputando valores ausentes na coluna "Review Rating" com a classificação mediana da categoria do produto
df ['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
# Renomeando colunas de acordo com o formato snake case para melhor legibilidade e documentação
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# Criando a coluna (age_group) com base na coluna idade (age)

labels = ['Young Adult', 'Adult', 'Middle_age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# criando uma nova coluna purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Monthly': 30,
    'every 3 months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropando a coluna promo_code_used
df = df.drop('promo_code_used', axis=1)
df.columns
```
